```python
# cette ligne est pour le notebook ipython, vous pouvez l'ignorer
%pylab inline

import netCDF4

# lecture des donnees
nc = netCDF4.Dataset('t2.2000.asm.nc')
lat = nc.variables['lat'][:]
lon = nc.variables['lon'][:]
t2 = nc.variables['t2'][:]

%whos
```

```python
plt.pcolor(lon, lat, t2_map)
plt.xlim(0, 360)
plt.ylim(-90, 90)
plt.colorbar()
```

L'image est tres jolie mais n'est pas une "vraie" carte -- pas de projection, pas moyen d'ajouter les continents, etc. Pour cela, nous allons utiliser le module basemap.

## 3. Intermediaire : zoom sur la France

On commence par une carte globale

```python
# importation du module Basemap
from mpl_toolkits.basemap import Basemap
```

```python
# creation d'une carte simple, sans specifier de projection
mymap = Basemap()
# conversion des lat/lon en coordonnees ecran
x, y = mymap(lon, lat)
# appel de la fonction pcolor de la carte
mymap.pcolor(x, y, t2_map)

# on rajoute les limites des continents
mymap.drawcoastlines()
```

Probleme ! On n'a que la moitie de la carte.

C'est parce que les longitudes de la carte sont entre 0 et 360 alors que Basemap attend des longitudes entre -180 et 180.
Heureusement, Basemap contient une fonction expres pour recaler les longitudes et la carte qui va avec.

```python
from mpl_toolkits.basemap import shiftgrid

# parametres de shiftgrid :
# newmap, newlon = shiftgrid(newlon0, map, lon)

# newlon0 = nouveau point de depart pour les longitudes (ici -180)
# map = la carte a recaler
# lon = les longitudes a recaler.

# il faut noter que shiftgrid veut que newlon0 existe quelque part dans lon. 
# Du coup, ici nous sommes obliges de decaler lon de 360 degres vers l'ouest.

print lon[0:5]
t2_map, lon = shiftgrid(-180., t2_map, lon-360)
print lon[0:5]
```

```python
# on refait la carte
x, y = mymap(lon, lat)
mymap.pcolor(x, y, t2_map)
mymap.drawcoastlines()
```

```python
# Comment zoomer sur la France ?
# il faut maintenant passer des parametres a la fonction Basemap en creant notre carte.
# llcrnrlon = lower-left corner lon, etc.

# on passe egalement un parametre 'resolution' qui indique la precision des donnees pour les limites de continents. 
# resolution peut etre 'c' (defaut, low-res), 'l' (low) 'i' (intermediate), 'h' (high).
# cf http://matplotlib.org/basemap/api/basemap_api.html

plt.figure(figsize=[14,8])
france = Basemap(llcrnrlon=-7.,llcrnrlat=41,urcrnrlon=12,urcrnrlat=51.5, resolution='i')
x, y = france(lon, lat)
france.pcolor(x, y, t2_map)

# indiquons le sirta pour rigoler : 48.71N, 2.21E
x, y = france(2.21, 48.71)
france.plot(x, y, marker='*', markersize=12)
# voir http://matplotlib.org/api/markers_api.html

# quelques fonctions de carte sympa
france.drawcoastlines()
france.drawcountries()

# le parametre labels precise ou indiquer les valeurs de meridiens et paralleles :
# [left,right,top,bottom]
# cf. http://matplotlib.org/basemap/users/graticule.html

france.drawmeridians(np.r_[-180:180:2.5], labels=[0,0,0,1])
france.drawparallels(np.r_[-180:180:2.5], labels=[1,0,0,0])

# utile pour certaines cartes, pas forcement ici
france.drawmapboundary()

# on change l'intervalle de couleurs affichées
plt.clim(275, 290)

# on ajoute une colorbar avec un titre
cb = plt.colorbar()
cb.set_label('Temperature [K]')
```

## 4. Anomalie de temperature pour 1976

* Afficher la carte d'anomalie de temperature pour 1976 compare a la periode 1970-2001

```python
# Maintenant il faut lire *toutes* les donnees de 1970 a 2001.
# pour recuperer les noms de fichier on utilise la fonction "glob" du module glob.

import glob
files = glob.glob('t2.*.asm.nc')
files.sort()  # pour les ranger en ordre
print files
```

```python
# lecture de toutes les donnees (peut prendre du temps...)

t2_liste = []
for f in files:
    nc = netCDF4.Dataset(f)
    t2 = nc.variables['t2'][:]
    t2_year = t2.mean(axis=0)
    t2_liste.append(t2_year)

# t2_liste contient une carte par an
print np.shape(t2_liste)
# on calcule la moyenne sur les annees
t2_average = np.mean(t2_liste, axis=0)
```

```python
# anomalie pour 1976
anom76 = t2_liste[6] - t2_average
print np.shape(anom76)
# on recale les longitudes
anom76, lon = shiftgrid(-180., anom76, lon-360)
```

```python
# on fait une carte d'anomalie
# on demande a pcolor une colormap sympa pour les anomalies
# cf http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps pour les colormaps disponibles

x, y = france(lon, lat)

plt.figure(figsize=[14,8])
france.pcolor(x, y, anom76, cmap = plt.get_cmap('RdBu_r'))
france.drawcoastlines()
france.drawcountries()
plt.clim(-1,1)
plt.title('Temperature Anomaly for 1976 vs. 1970-2000')
cb = plt.colorbar()
cb.set_label('Degrees')
```
