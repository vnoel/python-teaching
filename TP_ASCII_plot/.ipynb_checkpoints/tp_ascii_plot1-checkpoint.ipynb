L'affichage des donnees est pollué par la presence de valeurs a -999, symptomatiques de donnees invalides. On les prend en compte en les "masquant" dans le vecteur numpy cblow.

```python
cblow = np.ma.masked_where(cblow < 0, cblow)
cbmid = np.ma.masked_where(cbmid < 0, cbmid)
cbhigh = np.ma.masked_where(cbhigh < 0, cbhigh)
plt.plot(cblow)
```

Probleme : la commande hist ne comprend pas les valeurs masquées (-999), il faut les eliminer a la main.

```python
idx = (cblow >= 0)
plt.hist(cblow[idx])
```

```python
# Les trois distributions dans des subplots separes

# creation d'une fenetre figure plus large que haute
plt.figure(figsize=[15,5])

# trois subplots
plt.subplot(1,3,1)
plt.hist(cblow[idx] / 1000)
plt.title('Low-level clouds')
plt.xlabel('Altitude [km]')

plt.subplot(1,3,2)
plt.hist(cbmid[cbmid >= 0] / 1000)
plt.title('Mid-level clouds')
plt.xlabel('Altitude [km]')

plt.subplot(1,3,3)
plt.hist(cbhigh[cbhigh >= 0] / 1000)
plt.title('High-level clouds')
plt.xlabel('Altitude [km]')
```

La meme chose d'une facon un peu plus python, en utilisant des dictionnaires

```python
# creation d'un dictionnaire contenant les series temporelles nommees
altitudes = {'Low-level clouds':cblow, 'Mid-level clouds':cbmid, 'High-level clouds':cbhigh}

# creation d'une fenetre figure plus large que haute
plt.figure(figsize=[15,5])

# boucle sur les elements du dictionnaire
for i, cloud_type in enumerate(altitudes):
    plt.subplot(1,3,i+1)
    idx = (altitudes[cloud_type] >= 0)
    plt.hist(altitudes[cloud_type][idx] / 1000)
    plt.title(cloud_type)
    plt.xlabel('Altitude [km]')
```

On remarque au passage que les elements d'un dictionnaire ne sont pas ordonnes.

```python
# Affichage des trois distributions sur le meme graphe

# cette fois-ci on choisit les bins dans lesquels faire la distribution
bins = np.r_[0:14:0.5]

for i, cloud_type in enumerate(altitudes):
    idx = (altitudes[cloud_type] >= 0)
    plt.hist(altitudes[cloud_type][idx] / 1000, bins=bins, label=cloud_type)

plt.xlabel('Altitude [km]')
plt.legend()
```

```python
dates = []
for row in x:
    annee, mois, jour, heure, minute = row[0:5].astype('int')
    date = datetime(annee, mois, jour, heure, minute)
    dates.append(date)
print dates[0:10]
```

```python
# Affichage en utilisant les dates
fig = plt.figure(figsize=[15,5])
plt.plot(dates, cblow, 'blue', label='low-level clouds')
plt.plot(dates, cbmid, 'red', label='mid-level clouds')
plt.plot(dates, cbhigh, 'green', label='high-level clouds')
fig.autofmt_xdate()
plt.ylabel('Altitude [m]')
plt.legend()
```
